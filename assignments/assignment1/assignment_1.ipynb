Task 1: A list of dictionaries

```python
transactions = [
    {"id": 1, "product": "Laptop", "price": 1200, "quantity": 2, "region": "EU"},
    {"id": 2, "product": "Phone", "price": 800, "quantity": -1, "region": "US"},
    {"id": 3, "product": "Tablet", "price": 0, "quantity": 5, "region": "EU"},
    {"id": 4, "product": "Monitor", "price": 350, "quantity": 3, "region": "US"},
    {"id": 5, "product": "Laptop", "price": 1200, "quantity": 1, "region": "EU"},
    {"id": 6, "product": "Keyboard", "price": 75, "quantity": 10, "region": "APAC"},
]
```

```python
# A: Filters out invalid transactions

valid= [t for t in transactions if t["price"] > 0 and t["quantity"] > 0]

print(valid)
```

```python
# B: Add a total field

for t in valid:
    t["total"] = t["price"] * t["quantity"]
```

```python
# C: Grouping per region and adding total revenue per region

revenue_per_region = {}
for t in valid:
    region = t["region"]
    if region not in revenue_per_region:
        revenue_per_region[region] = 0
        revenue_per_region[region] += t["total"]
```

```python
# D: Print the value for the highest revenue per region

top= max(revenue_per_region, key=revenue_per_region.get)
print(f"Top region: {top} with {revenue_per_region[top]}")
```

Task 2: Creating a function


```python
sales = [1200, 980, 1540, 870, 2010, 1650, 1300, 1100, 950, 1780, 2200, 1430, 1600, 1900]
```

```python
def summarise_sales(daily_sales):
    total = sum(daily_sales)
    average = round(total / len(daily_sales), 2)
    best_day = daily_sales.index(max(daily_sales)) + 1
    worst_day = daily_sales.index(min(daily_sales)) + 1
    
    above_average_count = 0
    for sale in daily_sales:
        if sale > average:
            above_average_count += 1

    return {"total": total, "average": average, "best_day": best_day, "worst_day": worst_day, "above_average_count": above_average_count}


print(summarise_sales(sales))
```

Task 3: Loops

```python
students = [
    {"name": "Anna", "grade": 85, "attendance": 0.92},
    {"name": "Erik", "grade": 58, "attendance": 0.75},
    {"name": "Mia", "grade": 91, "attendance": 0.98},
    {"name": "Leo", "grade": 44, "attendance": 0.60},
    {"name": "Sara", "grade": 72, "attendance": 0.88},
    {"name": "Tom", "grade": 67, "attendance": 0.70},
    {"name": "Nina", "grade": 95, "attendance": 0.99},
    {"name": "Olli", "grade": 50, "attendance": 0.65},
]
```

```python
# A: creating a list that only includes students with a grade above 50 and printing it

passing = [s["name"] for s in students if s["grade"] >=50]

print(passing)
```

```python
# B: grading the students

for s in students:
    if s["grade"] >= 80:
        print(s["name"] + ": Excellent")
    elif s["grade"] >= 50:
        print(s["name"] + ": Pass")
    else:
        print(s["name"] + ": Fail")
```

```python
# C: How many students have an attendance below 0.80

count = 0
for s in students:
    if s["attendance"] < 0.80:
        count += 1

print("students with low attendance:", count)
```

```python
# D: Student with the highest grade

top_student = students[0]
for s in students:
    if s["grade"] > top_student["grade"]:
        top_student = s

print(top_student["name"], top_student["grade"])
```

```python
# E: Student with highest grade using max()

top_student = max(students, key=lambda s: s["grade"])

print(top_student["name"], top_student["grade"])
```

Task 4: Class and methods

```python
class Portfolio:
    def __init__(self, owner):
        self.owner = owner
        self.holdings = {}

    # A: adds a new asset or updates the number of shares if the ticker already exists 
    def add_asset(self, ticker, shares, price_per_share):
        if ticker in self.holdings:
            self.holdings[ticker]["shares"] += shares
            self.holdings[ticker]["price_per_share"] = price_per_share
        else:
            self.holdings[ticker] = {
                "shares": shares,
                "price_per_share": price_per_share
            }

    # B: Calculating total value of the portfolio
    def total_value(self):
        total = 0
        for ticker in self.holdings:
            total += self.holdings[ticker]["shares"] * self.holdings[ticker]["price_per_share"]
        return total

    # C: returns the ticker of the asset with the highest total value
    def most_valuable(self):
        top_ticker = None
        top_value = 0
        for ticker in self.holdings:
            value = self.holdings[ticker]["shares"] * self.holdings[ticker]["price_per_share"]
            if value > top_value:
                top_value = value
                top_ticker = ticker
        return top_ticker

    # D:  returns a readable summary of the portfolio, including the owner's name, the number of assets, and the total value
    def __str__(self):
        return (f"Owner: {self.owner}\n"
                f"Number of assets: {len(self.holdings)}\n"
                f"Total value: {self.total_value()}")


p = Portfolio("Anna")
p.add_asset("AAPL", 10, 189)
p.add_asset("MSFT", 5, 415)
p.add_asset("NVDA", 8, 875)

print(p)
print("Most valuable asset:", p.most_valuable())
print("Total value:", p.total_value())
```

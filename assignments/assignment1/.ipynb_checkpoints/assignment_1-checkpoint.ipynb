### Task 1: A list of dictionaries

```python
transactions = [
    {"id": 1, "product": "Laptop", "price": 1200, "quantity": 2, "region": "EU"},
    {"id": 2, "product": "Phone", "price": 800, "quantity": -1, "region": "US"},
    {"id": 3, "product": "Tablet", "price": 0, "quantity": 5, "region": "EU"},
    {"id": 4, "product": "Monitor", "price": 350, "quantity": 3, "region": "US"},
    {"id": 5, "product": "Laptop", "price": 1200, "quantity": 1, "region": "EU"},
    {"id": 6, "product": "Keyboard", "price": 75, "quantity": 10, "region": "APAC"},
]
```

```python
# A: Filters out invalid transactions

valid= [t for t in transactions if t["price"] > 0 and t["quantity"] > 0]

print(valid)
```

```python
# B: Add a total field

for t in valid:
    t["total"] = t["price"] * t["quantity"]
```

```python
# C: Grouping per region and adding total revenue per region

revenue_per_region = {}
for t in valid:
    region = t["region"]
    if region not in revenue_per_region:
        revenue_per_region[region] = 0
        revenue_per_region[region] += t["total"]
```

```python
# D: Print the value for the highest revenue per region

top= max(revenue_per_region, key=revenue_per_region.get)
print(f"Top region: {top} with {revenue_per_region[top]}")
```

### Task 2: Creating a function


```python
sales = [1200, 980, 1540, 870, 2010, 1650, 1300, 1100, 950, 1780, 2200, 1430, 1600, 1900]
```

```python
def summarise_sales(daily_sales):
    total = sum(daily_sales)
    average = round(total / len(daily_sales), 2)
    best_day = daily_sales.index(max(daily_sales)) + 1
    worst_day = daily_sales.index(min(daily_sales)) + 1
    
    above_average_count = 0
    for sale in daily_sales:
        if sale > average:
            above_average_count += 1

    return {"total": total, "average": average, "best_day": best_day, "worst_day": worst_day, "above_average_count": above_average_count}


print(summarise_sales(sales))
```

### Task 3: Loops

```python
students = [
    {"name": "Anna", "grade": 85, "attendance": 0.92},
    {"name": "Erik", "grade": 58, "attendance": 0.75},
    {"name": "Mia", "grade": 91, "attendance": 0.98},
    {"name": "Leo", "grade": 44, "attendance": 0.60},
    {"name": "Sara", "grade": 72, "attendance": 0.88},
    {"name": "Tom", "grade": 67, "attendance": 0.70},
    {"name": "Nina", "grade": 95, "attendance": 0.99},
    {"name": "Olli", "grade": 50, "attendance": 0.65},
]
```

```python
# A: creating a list that only includes students with a grade above 50 and printing it

passing = [s["name"] for s in students if s["grade"] >=50]

print(passing)
```

```python
# B: grading the students

for s in students:
    if s["grade"] >= 80:
        print(s["name"] + ": Excellent")
    elif s["grade"] >= 50:
        print(s["name"] + ": Pass")
    else:
        print(s["name"] + ": Fail")
```

```python
# C: How many students have an attendance below 0.80

count = 0
for s in students:
    if s["attendance"] < 0.80:
        count += 1

print("students with low attendance:", count)
```

```python
# D: Student with the highest grade

top_student = students[0]
for s in students:
    if s["grade"] > top_student["grade"]:
        top_student = s

print(top_student["name"], top_student["grade"])
```

```python
# E: Student with highest grade using max()

top_student = max(students, key=lambda s: s["grade"])

print(top_student["name"], top_student["grade"])
```

### Task 4: Class and methods

```python
class Portfolio:
    def __init__(self, owner):
        self.owner = owner
        self.holdings = {}

    # A: adds a new asset or updates the number of shares if the ticker already exists 
    def add_asset(self, ticker, shares, price_per_share):
        if ticker in self.holdings:
            self.holdings[ticker]["shares"] += shares
            self.holdings[ticker]["price_per_share"] = price_per_share
        else:
            self.holdings[ticker] = {
                "shares": shares,
                "price_per_share": price_per_share
            }

    # B: Calculating total value of the portfolio
    def total_value(self):
        total = 0
        for ticker in self.holdings:
            total += self.holdings[ticker]["shares"] * self.holdings[ticker]["price_per_share"]
        return total

    # C: returns the ticker of the asset with the highest total value
    def most_valuable(self):
        top_ticker = None
        top_value = 0
        for ticker in self.holdings:
            value = self.holdings[ticker]["shares"] * self.holdings[ticker]["price_per_share"]
            if value > top_value:
                top_value = value
                top_ticker = ticker
        return top_ticker

    # D:  returns a readable summary of the portfolio, including the owner's name, the number of assets, and the total value
    def __str__(self):
        return (f"Owner: {self.owner}\n"
                f"Number of assets: {len(self.holdings)}\n"
                f"Total value: {self.total_value()}")


p = Portfolio("Anna")
p.add_asset("AAPL", 10, 189)
p.add_asset("MSFT", 5, 415)
p.add_asset("NVDA", 8, 875)

print(p)
print("Most valuable asset:", p.most_valuable())
print("Total value:", p.total_value())
```

### Task 5: Importing a module

```python
orders = [
    ("2025-01-15", "Laptop"), ("2025-01-22", "Phone"),
    ("2025-02-03", "Laptop"), ("2025-02-14", "Tablet"),
    ("2025-03-01", "Phone"), ("2025-03-10", "Phone"),
    ("2025-03-22", "Laptop"), ("2025-04-05", "Monitor"),
]
```

```python
from datetime import date

orders = [
    ("2025-01-15", "Laptop"), ("2025-01-22", "Phone"),
    ("2025-02-03", "Laptop"), ("2025-02-14", "Tablet"),
    ("2025-03-01", "Phone"), ("2025-03-10", "Phone"),
    ("2025-03-22", "Laptop"), ("2025-04-05", "Monitor"),
]
```

```python
# coverting text dates into date format

order_dates = [date.fromisoformat(order_date) for order_date, _ in orders]
```

```python
# Computing and printing the number of days between the earliest and latest order

earliest = min(order_dates)
latest = max(order_dates)
days_between = (latest - earliest).days

print("Earliest order:", earliest)
print("Latest order:", latest)
print("Days between:", days_between)
```

### Task 6: Array Operations and Broadcasting

```python
# Rows: years (2022, 2023, 2024), Columns: quarters (Q1, Q2, Q3, Q4)

import numpy as np #This loads numpy and I can refer to it as np later

revenue = np.array([
    [210, 230, 190, 250],
    [220, 260, 200, 270],
    [250, 290, 230, 310]])
```

```python
# A:  Compute and print the total annual revenue for each year

annual_totals = revenue.sum(axis=1)

annual_totals = revenue.sum(axis=1)
print("Annual totals:")
print(f"  2022: {annual_totals[0]}")
print(f"  2023: {annual_totals[1]}")
print(f"  2024: {annual_totals[2]}")
```

```python
# B: Compute the mean quarterly revenue across all years

quarterly_means = revenue.mean(axis=0)

quarterly_means = revenue.mean(axis=0)
print("Quarterly means:")
print(f"  Q1: {quarterly_means[0]:.2f}")
print(f"  Q2: {quarterly_means[1]:.2f}")
print(f"  Q3: {quarterly_means[2]:.2f}")
print(f"  Q4: {quarterly_means[3]:.2f}")
```

```python
# C: Using broadcasting, apply a growth adjustment to the 2024 row

growth = np.array([1.05, 1.03, 1.08, 1.02])
adjusted_2024 = revenue[2] * growth

growth = np.array([1.05, 1.03, 1.08, 1.02])
adjusted_2024 = revenue[2] * growth
print("Adjusted 2024 revenue:")
print(f"  Q1: {adjusted_2024[0]}")
print(f"  Q2: {adjusted_2024[1]}")
print(f"  Q3: {adjusted_2024[2]}")
print(f"  Q4: {adjusted_2024[3]}")
```

```python
# D: Use boolean indexing to find and print all quarterly revenue values that exceed 250

above_250 = revenue[revenue > 250]
print("Values above 250:", above_250)
```

### Task 7: Pandas

```python
import pandas as pd

data = {
    "employee":          ["Anna", "Erik", "Mia", "Leo", "Sara", "Tom"],
    "department":        ["Finance", "Marketing", "Finance", "IT", "Marketing", "IT"],
    "salary":            [4500, 4000, 4800, 5200, 3900, 5100],
    "years_exp":         [5, 3, 7, 10, 2, 8],
    "performance_score": [4.2, 3.8, 4.5, 4.0, 3.5, 4.9]
}
df = pd.DataFrame(data)
print(df)
```

```python
# A: Add a newc column

df["salary_per_year_exp"] = round(df["salary"] / df["years_exp"])

print(df)
```

```python
# B: sort and print top3 employees

top3= df.sort_values("performance_score", ascending=False).head(3)

print(top3)
```

```python
# C: all employees in the Finance department with a performance score above 4.0

finance_high = df.loc[
    (df["department"] == "Finance") & (df["performance_score"] > 4.0)
    ]

print(finance_high)
```

```python
# D: renaming a column

df = df.rename(columns={"years_exp": "experience_years"})
print(df.columns)
```

```python
# E:  select and print the salary and performance score columns for the first four employees

subset = df.iloc[:4][["salary", "performance_score"]]

print(subset)
```

### Task 8: More Pandas

```python
import numpy as np
import pandas as pd

rng = np.random.default_rng(seed=123)
days = pd.date_range(start="2025-01-06", periods=28, freq="D")
visitors = rng.integers(200, 600, size=28)
conversion_rate = rng.uniform(0.02, 0.08, size=28).round(4)
avg_order_value = rng.normal(50, 10, size=28).round(2)
```

```python
# A: creating the data frame

traffic = pd.DataFrame({
    "visitors": visitors,
    "conversion_rate": conversion_rate,
    "avg_order_value": avg_order_value
}, index=days)

print(traffic)
```

```python
# B: Computing two new columns

traffic["conversions"] = round(traffic["visitors"] * traffic["conversion_rate"])
traffic["daily_revenue"] = traffic["conversions"] * traffic["avg_order_value"]

print(traffic)
```
